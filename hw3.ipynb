```python
import numpy as np
import matplotlib.pyplot as plt


data = np.loadtxt('hw3Data-3\hw3Data\D2z.txt')

def dd(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))


def nn(train, test):
    min = float('inf')
    pre = None
    for d in train:
        dis = dd(d[:2], test)
        if dis < min:
            min = dis
            pre = d[2]
    return pre

grid = np.mgrid[-2:2:0.1, -2:2:0.1]

grid = grid.reshape(2, -1)

grid = grid.T


testY = []
for point in grid:
    testY.append(nn(data, point))

plt.scatter(data[data[:,2] == 0][:,0], data[data[:,2] == 0][:,1], c='blue' )
plt.scatter(data[data[:,2] == 1][:,0], data[data[:,2] == 1][:,1], c='red')


for i, prediction in enumerate(testY):
    if prediction == 0:
        plt.scatter(grid[i,0], grid[i,1],c='blue',  s=10)
    else:
        plt.scatter(grid[i,0], grid[i,1],c='red',  s=10)

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score

df = pd.read_csv('hw3Data-3\hw3Data\emails.csv')

X = df.drop(columns=["Email No.",'Prediction'])
y = df['Prediction']



def nn(k,q=0):
    knn = KNeighborsClassifier(n_neighbors=k)

    kf = KFold(n_splits=5)
    
    avg = 0

    for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        knn.fit(X_train, y_train)

        y_pred = knn.predict(X_test.values)

        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)

        if(q==1):
            print(f"Fold {fold+1}:")
            print(f"Accuracy: {accuracy:.4f}")
            print(f"Precision: {precision:.4f}")
            print(f"Recall: {recall:.4f}")
            print()
            
        avg = avg + accuracy
        
    return avg/5
```
